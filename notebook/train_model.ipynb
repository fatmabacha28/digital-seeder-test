# Entraînement du Modèle de Classification de Documents

Ce notebook démontre comment entraîner le modèle scikit-learn pour classer les textes extraits des PDF en 3 catégories : `BUSINESS`, `ENTERTAINMENT`, et `AUTRE`.

```python
import pandas as pd
import joblib
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
```

## 1. Préparation des données (Mock)
Dans un cas réel, ces données proviendraient de l'extraction de centaines de PDF.

```python
data = {
    'text': [
        'bilan financier 2023 compte de résultat entreprise',
        'contrat de prestation de service client',
        'facture achat matériel informatique',
        'invitation avant première film cinéma grand rex',
        'ticket concert pass festival musique été',
        'pièce théâtre comédie mise en scène',
        'recette de cuisine gâteau au chocolat',
        'manuel utilisation aspirateur robot',
        'ordonnance médicale docteur pharmacie'
    ],
    'category': [
        'BUSINESS',
        'BUSINESS',
        'BUSINESS',
        'ENTERTAINMENT',
        'ENTERTAINMENT',
        'ENTERTAINMENT',
        'AUTRE',
        'AUTRE',
        'AUTRE'
    ]
}

df = pd.DataFrame(data)
df.head()
```

## 2. Entraînement du modèle

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['category'], test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

## 3. Évaluation

```python
y_pred = model.predict(X_test_vec)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))
```

## 4. Sauvegarde des artefacts pour l'API Flask

```python
models_dir = '../models'
os.makedirs(models_dir, exist_ok=True)

joblib.dump(model, os.path.join(models_dir, 'model.joblib'))
joblib.dump(vectorizer, os.path.join(models_dir, 'vectorizer.joblib'))
print("Modèles sauvegardés avec succès dans le dossier 'models'.")
```
